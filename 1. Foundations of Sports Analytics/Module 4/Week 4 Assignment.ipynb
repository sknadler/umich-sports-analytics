## Part 1 - Import and Explore Data 

```python
#Import Libraries
import pandas as pd
import numpy as np
import datetime
import statsmodels.formula.api as sm
```

```python
#Import NFL_Game dataset
NFL_Game=pd.read_csv("Assignment Data/Week 4/nfl_game.csv")
display(NFL_Game)
```

```python
# Display summary statistics for date

NFL_Game['date'].describe()
```

```python
# Display summary statistics for score based on home or away game

NFL_Game.groupby(['home']).describe()
```

```python
# Determine correlation coefficients for the following pairs of variables

print(NFL_Game['win'].corr(NFL_Game['home']))
print(NFL_Game['score_diff'].corr(NFL_Game['home']))
print(NFL_Game['score'].corr(NFL_Game['weather_temperature']))
print(NFL_Game['score'].corr(NFL_Game['weather_humidity']))
print(NFL_Game['score'].corr(NFL_Game['weather_wind_mph']))
```

## Part 2 - Regression Analysis 1 - Test of Home Game Advantage

```python
# Run regression on score differential based only on home or away game

reg1_1 = sm.ols(formula = 'score_diff ~ home', data = NFL_Game).fit()
reg1_1.summary()
```

```python
# Run regression on score differential based on home or away, stadium capacity, and stadium neutrality

reg1_2 = sm.ols(formula = 'score_diff ~ home+stadium_capacity+stadium_neutral+home*stadium_neutral', data = NFL_Game).fit()
reg1_2.summary()
```

```python
# Run regression on score differential based on home or away, stadium capacity, stadium neutrality, team, and opponent

reg1_2 = sm.ols(formula = 'score_diff ~ home+stadium_capacity+stadium_neutral+home*stadium_neutral+team+opponent', data = NFL_Game).fit()
reg1_2.summary()
```

## Part 3 - Regression Analysis 2 -- Impact of Outside Factors on Scores

```python
# Run regression on score based on season and home or away game

reg2_1 = sm.ols(formula = 'score ~ season+home', data = NFL_Game).fit()
reg2_1.summary()
```

```python
# Run regression on score based on season, home or away, temperature, wind speed, and humidity

reg2_2 = sm.ols(formula = 'score ~ season+home+weather_temperature+weather_wind_mph+weather_humidity', data = NFL_Game).fit()
reg2_2.summary()
```

```python
# Run regression on score based on season, home or away, weather, and stadium information

reg2_3 = sm.ols(formula = 'score ~ season+home+weather_temperature+weather_wind_mph+weather_humidity+stadium_capacity+stadium_age+stadium_type+stadium_neutral+home*stadium_neutral', data = NFL_Game).fit()
reg2_3.summary()
```

```python
# Run regression on score based on season, home or away, weather, stadium information, team, and opponent

reg2_4 = sm.ols(formula = 'score ~ season+home+weather_temperature+weather_wind_mph+weather_humidity+stadium_capacity+stadium_age+stadium_type+stadium_neutral+home*stadium_neutral+team+opponent', data = NFL_Game).fit()
reg2_4.summary()
```
