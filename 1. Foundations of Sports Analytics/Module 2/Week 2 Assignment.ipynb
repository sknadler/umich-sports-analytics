## Part 1- Data Coding and Merging

```python
#Import Libraries

import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import NHL Team Data and Display
NHL_Team=pd.read_csv("Assignment Data/Week 2/NHL_team.csv")
display(NHL_Team)
```

```python
#Import NHL Competition Data and Display Head
NHL_Competition=pd.read_csv("Assignment Data/Week 2/NHL_competition.csv")
NHL_Competition.head()
```

```python
#Import NHL Game Data and Display Head
NHL_Game=pd.read_csv("Assignment Data/Week 2/NHL_game.csv")
NHL_Game.head()
```

```python
#Import NHL PPPK Data and Display Head
NHL_PPPK=pd.read_csv("Assignment Data/Week 2/pp.pk.ppgf.csv")
NHL_PPPK.head()
```

```python
# Clean NHL_Team dataframe by deleting and renaming columns

NHL_Team.drop(columns = ['Unnamed: 0', 'abbr', 'tname', 'lname', 'sname'], inplace = True)
NHL_Team.rename(columns = {'name': 'team_name'}, inplace = True)
NHL_Team.head()
```

```python
# Clean NHL_Competition dataframe by deleting and renaming columns

NHL_Competition.drop(columns = ['Unnamed: 0', 'tz', 'start', 'end'], inplace = True)
NHL_Competition.rename(columns = {'name': 'competition_name'}, inplace = True)
NHL_Competition.head()
```

```python
# Clean NHL_Game dataframe by deleting columns

NHL_Game.drop(columns = ['X', 'period', 'status'], inplace = True)
NHL_Game.head()
```

```python
# Merge NHL_Team into NHL_Game by 'tid'

NHL_Game = pd.merge(NHL_Team, NHL_Game, on = ['tid'])
NHL_Game.head()
```

```python
# Merge NHL_Game into NHL_Competition by 'comp_id'

NHL_Game = pd.merge(NHL_Competition, NHL_Game, on = ['comp_id'])
NHL_Game.head()
```

```python
# Keep track of goal differentials for each game

NHL_Game['hgd'] = NHL_Game['hscore'] - NHL_Game['ascore']
NHL_Game = NHL_Game[pd.notnull(NHL_Game['hgd'])]
```

```python
NHL_Game.dropna()
NHL_Game.head()
```

```python
# Convert game dates to datetime type

import datetime
NHL_Game['date'] = pd.to_datetime(NHL_Game['date'])
NHL_Game['date'].head()
```

```python
# Sort games by date

NHL_Game.sort_values(by = ['date'], inplace = True)
NHL_Game.head(15)
```

```python
# Separate home games and away games for each team

NHL_Home = NHL_Game[NHL_Game['home_away'] == 'home']
NHL_Away = NHL_Game[NHL_Game['home_away'] == 'away']
```

```python
# Rename variables in NHL_Home and NHL_Away

NHL_Home.rename(columns = {'hscore': 'goals_for', 'ascore': 'goals_against'}, inplace = True)
NHL_Away.rename(columns = {'hscore': 'goals_against', 'ascore': 'goals_for'}, inplace = True)
```

```python
# Keep track of wins per team

NHL_Home['win'] = np.where(NHL_Home['goals_for'] > NHL_Home['goals_against'], 1, np.where(NHL_Home['goals_for'] < NHL_Home['goals_against'], 0, 0.5))
NHL_Away['win'] = np.where(NHL_Away['goals_for'] > NHL_Away['goals_against'], 1, np.where(NHL_Away['goals_for'] < NHL_Away['goals_against'], 0, 0.5))
```

```python
# Merge dataframes back on NHL_Game

NHL_Game = NHL_Home.append(NHL_Away)
NHL_Game.head()
```

```python
# Aggregate total wins, goals for, and goals against for each team

NHL_Team_Stats = NHL_Game.groupby(['tid', 'competition_name', 'type'])['win', 'goals_for', 'goals_against'].sum().reset_index()
NHL_Team_Stats
```

```python
# Create dataframe of game type distributions for each team

NHL_Game['game_count'] = 1
NHL_Game_Count = NHL_Game.groupby(['tid', 'competition_name', 'type'])['game_count'].sum().reset_index()
NHL_Game_Count
```

```python
# Merge NHL_Team_Stats on NHL_Game_Count

NHL_Team_Stats = pd.merge(NHL_Team_Stats, NHL_Game_Count, on = ['tid', 'competition_name', 'type'])
NHL_Team_Stats.head()
```

```python
# Merge NHL_Team on NHL_Team_Stats

NHL_Team_Stats = pd.merge(NHL_Team_Stats, NHL_Team, on = 'tid')
NHL_Team_Stats.head()
```

```python
# Merge NHL_PPPK on NHL_Team_Stats by 'tricode' and 'competition_name'

NHL_Team_Stats = pd.merge(NHL_Team_Stats, NHL_PPPK, on = ['tricode', 'competition_name'])
NHL_Team_Stats.head()
```

```python
# Create counters for wins, goals for, and goals against per game

NHL_Team_Stats['win_pct'] = NHL_Team_Stats['win'] / NHL_Team_Stats['game_count']
NHL_Team_Stats['avg_gf'] = NHL_Team_Stats['goals_for'] / NHL_Team_Stats['game_count']
NHL_Team_Stats['avg_ga'] = NHL_Team_Stats['goals_against'] / NHL_Team_Stats['game_count']
NHL_Team_Stats.head()
```

```python
NHL_Team_R_Stats = NHL_Team_Stats[NHL_Team_Stats['type'] == 2]
NHL_Team_R_Stats.head()
```

## Part 3 - Correlation Analyses


```python
# Create regression plot for goals for vs. team win percentage

sns.regplot(x = 'goals_for', y = 'win_pct', data = NHL_Team_R_Stats,  marker = '.')
plt.xlabel('Total goals for')
plt.ylabel('Win percentage')
plt.title('Relationship between Goals for and Winning Percentage', fontsize = 11)
```

```python
# Calculate correlation coefficient of the above graph

NHL_Team_R_Stats['goals_for'].corr(NHL_Team_R_Stats['win_pct'])
```

```python
# Create regression plot for goals for vs. team win percentage by season

sns.lmplot(x = 'goals_for', y = 'win_pct', hue = 'competition_name', data = NHL_Team_R_Stats)
plt.xlabel('Total goals for')
plt.ylabel('Win percentage')
plt.title('Relationship between Goals for and Winning Percentage', fontsize = 11)
```

```python
# Remove 2011 and 2012 NHL season data

NHL_Team_R_Stats = NHL_Team_R_Stats[~NHL_Team_R_Stats['competition_name'].isin(['2011 NHL Regular Season', '2012 NHL Regular Season'])]
```

```python
# Create regression plot for goals for vs. team win percentage without outliers

sns.regplot(x = 'goals_for', y = 'win_pct', data = NHL_Team_R_Stats,  marker = '.')
plt.xlabel('Total goals for')
plt.ylabel('Win percentage')
plt.title('Relationship between Goals for and Winning Percentage', fontsize = 11)
```

```python
# Calculate correlation coefficient of the above graph

NHL_Team_R_Stats['goals_for'].corr(NHL_Team_R_Stats['win_pct'])
```
